```python
# Import shared chat helper functions
from helper_functions import add_user_message, add_assistant_message, chat


messages = []

system_prompt = """
You are a patient math tutor.
You will help the student solve the problem step by step, and explain each step in detail.
You will not give the final answer until the student has solved the problem themselves.
You do not give any information that is NOT related to Mathematics and respectfully decline to answer any questions that are not related to Mathematics.
"""


user_message = "how do i solve the equation x^2 + 2x - 3 = 0?"
print(user_message)
add_user_message(messages, user_message)

answer = chat(messages, system=system_prompt)
add_assistant_message(messages, answer)
print(answer);


# Add Random question that is not related to math to test the system's ability to decline answering non-math questions
user_message = "what is the capital of USA?"
print(user_message)
add_user_message(messages, "what is capital of USA?")

answer = chat(messages, system=system_prompt)
add_assistant_message(messages, answer)
print(answer)

# Now what happens if i change system prompt??
system_prompt = """
You only know the language "GROOT" from movie Avengers and you can only say "I AM GROOT".
"""
user_message = "Ok then how do i solve the equation x^2 + 2x - 3 = 0?"
print(user_message)
add_user_message(messages, user_message)

answer = chat(messages, system=system_prompt)
add_assistant_message(messages, answer)
print(answer)
```
