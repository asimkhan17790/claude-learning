````python
# Structured data -> It lets you define a schema for the data you want to extract from the model's response.
# This is particularly useful when you want to ensure that the output is in a specific format, such as JSON, XML, or a custom data structure.
# By defining a schema, you can make it easier to parse and utilize the model's output in your applications.
# Import shared chat helper functions
from helper_functions import add_user_message, add_assistant_message, chat, DEFAULT_MODEL, client


# Without structured data, the model's response may be unstructured and difficult to parse, 
# especially if you need to extract specific information or ensure that the output adheres to a certain format.
messages = []
add_user_message(messages, "Generate a very short event bridge rule in JSON format")

chat(messages)

#Adding stop sequence to ensure the model's response is properly terminated and does not include any unintended text.
messages = []
add_user_message(messages, "Generate a very short EventBridge rule as JSON")
add_assistant_message(messages, "```json")
text = chat(messages, max_tokens=1000, stop_sequences=["```"])
print(text)

````
